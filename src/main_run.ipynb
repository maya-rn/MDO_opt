```python
# Starting variables
wingspan = 1.0 # 1 m
mid_chord = 0.15 # 15 cm
tip_chord = 0.15 # 15 cm
t_skin = 0.001 # 1 mm
root_twist = 0 # 0 deg
mid_twist = 0 # 0 deg
tip_twist = 0 # 0 deg
skin_mat = 0
spar_mat = 0

# # Hard-code these values into cost equation
# spar_volume = 0.303*0.15*0.05*wingspan
# S = wing_area(wingspan, mid_chord, tip_chord)
# skin_volume = 2*S*t_skin
# V_maxR, Preq, _, _ = aero(wingspan, mid_chord, tip_chord, root_twist, mid_twist, tip_twist, spar_mat, skin_mat, spar_volume, skin_volume, 0, 0)
# desired_range = range(V_maxR, Preq)
# desired_cost = price(spar_mat, skin_mat, spar_volume, skin_volume)

```

```python
func_inputs = [wingspan, mid_chord, tip_chord, t_skin, root_twist, mid_twist, tip_twist, skin_mat, spar_mat]
```

```python
def fitness_func(ga_instance, solution, solution_idx):
    fitness = cost_func(solution[0], solution[1], solution[2], solution[3], solution[4], solution[5], solution[6], solution[7], solution[8])
    return fitness
```

```python
gene_space = [
    {'low':0.3, 'high':100}, # wingspan
    {'low':0.001, 'high':1}, # mid chord
    {'low':0.001, 'high':1}, # tip chord
    {'low':1e-5, 'high':10}, # skin thickness
    {'low':-5, 'high':0}, # root twist
    {'low':-5, 'high':0}, # mid twist
    {'low':-5, 'high':0}, # tip twist
    [0,1,2,3], # skin material
    [0,1,2,3] # spar material
]

gene_type = [
    float, # wingspan
    float, # mid chord
    float, # tip chord
    float, # skin thickness
    float, # root twist
    float, # mid twist
    float, # tip twist
    int, # skin material
    int # spar material
]
```

```python
ga_inst = pygad.GA(num_generations=20,
                  num_parents_mating=5,
                  fitness_func=fitness_func,
                  sol_per_pop=10,
                  num_genes=len(func_inputs),
                  parent_selection_type="sss",
                  mutation_type="random",
                  mutation_percent_genes=10,
                  gene_space=gene_space,
                  gene_type=gene_type)

ga_inst.run()

ga_inst.plot_fitness()
```
